```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data
data = pandas.read_csv('data.csv')
data.head()
```

```python
plt.scatter(data.x1, data.x2, c=data.y)
plt.grid(True)
```

```python
# Split the data into X and y
X = numpy.array(data[['x1', 'x2']])
y = numpy.array(data['y'])
```

```python
# import statements for the classification algorithms
from sklearn.linear_model import LogisticRegression # Logistic Regression
from sklearn.neural_network import MLPClassifier # Neural Networks
from sklearn.tree import DecisionTreeClassifier # Decision Trees
from sklearn.ensemble import GradientBoostingClassifier # Gradient Boosting Classifier
from sklearn.svm import SVC # Support Vector Machines
```

```python
# Logistic Regression
clf = LogisticRegression()
clf.fit(X,y)
```

```python
xx, yy = np.mgrid[-5:5:.01, -5:5:.01]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = clf.predict_proba(grid)[:, 1].reshape(xx.shape)
f, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(xx, yy, probs, 25, cmap="RdBu",
                      vmin=0, vmax=1)
ax_c = f.colorbar(contour)
ax_c.set_label("$P(y = 1)$")
ax_c.set_ticks([0, .25, .5, .75, 1])
ax.scatter(X[100:,0], X[100:, 1], c=y[100:], s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1);
ax.scatter(data.x1, data.x2, c=data.y);
ax.set(aspect="equal",
       xlim=(-5, 5), ylim=(-5, 5),
       xlabel="$X_1$", ylabel="$X_2$");
```

```python
# Decision Tree Classifier
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```python
# Support Vector Machine Classifier
clf = SVC()
clf.fit(X,y)
```

```python
# Import statement for train_test_split
from sklearn.cross_validation import train_test_split

# Use the train_test_split function to split the data into training and testing sets.
# The size of the testing set should be 20% of the total size of the data.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```
